```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

import numpy as np

%matplotlib inline
```

```python
wom = pd.read_csv('womens_gamestats.csv')
men = pd.read_csv('mens_gamestats.csv')
```

```python
wom["gender"] = 'f'
men["gender"] = 'm'
df = wom.append(men, ignore_index=True)
df.head()
```

```python
#let's graph the top assisters vs. their turns
top_assists = df[["name", "team", "goals", "assists", "ds", "turns", "gender"]].groupby(["name", "gender", "team"], as_index=False).sum()
top_assists = top_assists[top_assists["assists"] > 13].sort_index(by="assists", ascending=False)
top_assists.head()
```

```python
g = sns.lmplot("assists", "turns", top_assists, scatter_kws={"marker":""}, size=9, hue="gender", col="gender")
ax = g.axes[0][0]
for name, stats in top_assists[top_assists["gender"] == "f"].iterrows():
    ax.annotate(stats["name"], xy=(stats["assists"], stats["turns"]), horizontalalignment='center')
ax.set_xlabel("Assists")
ax.set_ylabel("Turnovers")
ax.set_title("Females")

ax2 = g.axes[0][1]
for name, stats in top_assists[top_assists["gender"] == "m"].iterrows():
    ax2.annotate(stats["name"], xy=(stats["assists"], stats["turns"]), horizontalalignment='center')
ax2.set_xlabel("Assists")
ax2.set_title("Males")
```

```python
#How do we get a count of games alongside a summary of # of games?
tdf = df.copy()
tdf["game_count"] = 1
pergame = (tdf[["name", "team", "goals", "assists", "ds", "turns", "gender", "game_count"]]
     .groupby(["name", "team", "gender"], as_index=False))
pergame = pergame.agg({
    'goals': np.sum,
    'assists': np.sum,
    "ds": np.sum,
    "turns": np.sum,
    "game_count": "count",
})
pergame["goalspg"] = pergame["goals"] / pergame["game_count"]
pergame["turnspg"] = pergame["turns"] / pergame["game_count"]
pergame.head()
```

```python
tp = (pergame.groupby(["team", "game_count", "gender"], as_index=False)
 .sum()
 .sort("goalspg", ascending=False))
tp.head()
```

```python
g = sns.lmplot("goalspg", "turnspg", tp, scatter_kws={"marker":""}, size=9, hue="gender", col="gender")
ax = g.axes[0][0]
for idx, stats in tp[tp["gender"] == "f"].iterrows():
    ax.annotate(stats["team"], xy=(stats["goalspg"], stats["turnspg"]), horizontalalignment='center')
ax.set_xlabel("Goals per game")
ax.set_ylabel("Turnovers per game")
ax.set_title("Women's division")

ax2 = g.axes[0][1]
for idx, stats in tp[tp["gender"] == "m"].iterrows():
    ax2.annotate(stats["team"], xy=(stats["goalspg"], stats["turnspg"]), horizontalalignment='center')
ax2.set_xlabel("Goals per game")
ax2.set_title("Men's division")
```
